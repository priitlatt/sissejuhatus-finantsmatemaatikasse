#### **Monte Carlo meetod** on statistikal põhinev probleemide lahendamise meetod. ####

* Kui meil on ette antud mingi tõenäosus $P$, mis näitab mingitel tingimustel vea esinemist, on arvutite abil võimalik selliseid tingimusi korduvalt genereerida.
* Sündmuse esinemiste arvu jagamisel tingumuste genereerimise arvuga, peaksime saama ligikaudu tõenäosuse $P$.

# $\pi$ väärtuse arvutamine #

Kui meil on ring raadiusega $R$ sisestatud ruutu küljepikkusega $2R$, siis 
* ringi pindala on $S_{ring} = \pi R^2$
* ruudu pindala on $S_{ruut} = 4R^2$.

Niisiis ringi ja ruudu pindalad suhtuvad teineteisse kujul

$$ \frac{S_{ring}}{S_{ruut}} = \frac{\pi}{4} $$

Teisi sõnu, kui me valime $N$ punkti ruudu seest, peaks umbes

$$ \frac{N\pi}{4} $$

neist punktidest langema ringi sisse.

Olgu ringi raadius $R$.

* valime juhuslikult ruudu seest mingi punkti $(x, y)$

* kontrollime kas punkt $(x, y)$ asub ka ringis, see tähendab, kas $x^2 + y^2 < R^2$

* jätame meelde genereeritud punktide arvu $N$

* ja ringi sisse sattnud punktide arvu $M$

$\pi$ saab siis ligikaudu arvutada kujul:

$$ \pi \approx \frac{4M}{N} $$

Defineerime konstandi `PI` väärtuseks $3.14159265359$.

```python
PI = 3.14159265359
```

```python
def approximate_pi(times):
    in_x, in_y, out_x, out_y = [], [], [], []
    for _ in range(times):
        x, y = random(), random()
        if  np.hypot(x, y) < 1:
            in_x.append(x)
            in_y.append(y)
        else:
            out_x.append(x)
            out_y.append(y)
    return in_x, in_y, out_x, out_y
```

```python
x = np.arange(0, 1, 0.01)

times = (10, 100, 500, 1000, 5000, 10000, 20000, 50000)
rows = int(len(times)/4) + len(times) % 4
columns = 4
```

```python
plt.figure(figsize=(20, rows*5))
plt.rcParams['font.size'] = 10

results = []

for i, N in enumerate(times):
    in_x, in_y, out_x, out_y = approximate_pi(N)
    plt.subplot(rows, columns, i+1)
    plt.plot(in_x, in_y, 'go')
    plt.plot(out_x, out_y, 'ro')
    plt.plot(x, np.sqrt(1 - x**2), 'r--')
    plt.title(r"$N = %d$" % N)
    
    pi = 4.0*len(in_x) / N
    results.append((N, pi))
```

```python
for N, pi in results:
    print "n = %d: pi = %f (viga: %f)" % (N, pi, np.abs(PI-pi))
```

### Monte Carlo meetodite omadusi ###

* Sageli kasulik lahendamaks probleeme rakendusmatemaatikast või füüsikast;
* mõistlik kui probleemi ei saa analüütiliselt lahendada;
* üldiselt aeglane ja ebaefektiivne viis.

Viimast illustreeris ka $\pi$ arvutamine.

Defineerime eelnevaga analoogilise, kuid mõnevõrra efektiivsema funktsiooni $\pi$ ligikaudse väärtuse leidmiseks.

```python
def approximate_pi_2(N):
    M = 0
    for _ in range(N):
        if  np.hypot(random(), random()) < 1:
            M += 1
    return 4.0*M / N
```

```python
for N in (10, 100, 1000, 10000, 100000, 1000000, 10000000):
    print "\nN = %d" % N
    pi_1, pi_2 = approximate_pi_2(N), approximate_pi_2(N)
    print "pi = %.16f (viga %.16f)" % (pi_1, np.abs(PI-pi_1))
    print "pi = %.16f (viga %.16f)" % (pi_2, np.abs(PI-pi_2))
```

Uurime arvutuste käigus tekkivaid vigu.

```python
errors = np.array([np.abs(approximate_pi_2(10000) - PI) for _ in range(100)])
mean = errors.mean()
print "Vigade keskväärtus:", mean
print "Minimaalne viga:", errors.min()
print "Maksimaalne viga:", errors.max()

plt.figure(figsize=(10, 5))
plt.plot(sorted(errors), 'o')
plt.plot((0, 100), (mean, mean))
plt.show()
```

### Ühe muutuja funktsioonide integreerimine ###

Vaatleme vahemikus $(a, b)$ integreeruvat funktsiooni $f \colon X \rightarrow \mathbb{R}$, kus $X \subset \mathbb{R}$. Kui $x_1, \dots, x_N$ ühtlaselt jaotatud juhuslikud arvud vahemikust $(a, b)$, siis

$$ S = \frac{b - a}{n} \sum_{i=1}^{N} f(x_i) $$

on ligikaudne väärtus integraalile

$$ \int_a^b f(x)\ \mathrm{d}x. $$

Sellist meetodit kutsutakse tavaliselt **Monte Carlo integreerimiseks**.

```python
def integrate(func, a, b, N=1000):
    r = a + (b - a)*random(N)
    s = np.sum(func(r))
    return s*(b-a)/N
```

```python
x = np.arange(0, np.pi, 0.01)
plt.plot(x, np.sin(x))

Math(r'\int_{0}^{\pi} \sin(x)\ \mathrm{d}x \approx %.16f' % integrate(np.sin, 0.0, np.pi))
```

```python
x = np.arange(0, np.pi, 0.01)
plt.plot(x, np.cos(x))

Math(r'\int_{0}^{\pi} \cos(x)\ \mathrm{d}x \approx %.16f' % integrate(np.cos, 0.0, np.pi))
```

```python
x = np.arange(2, 3, 0.01)
plt.plot(x, np.log(x))

Math(r'\int_{2}^{3} \ln x\ \mathrm{d}x \approx %.16f' % integrate(np.log, 2.0, 3.0))
```

```python
x = np.arange(-1, 1, 0.01)
plt.plot(x, np.exp(x))

Math(r'\int_{-1}^{1} e^x\ \mathrm{d}x \approx %.16f' % integrate(np.exp, -1, 1))
```

### Mitme muutuja funktsioonide integreerimine ###

Vaatleme piirkonnas $\Omega$ integreeruvat funktsiooni $f \colon X \rightarrow \mathbb{R}$, kus $X \subset \mathbb{R}^m$. 

Meie eesmärgiks on arvutada $m$-muutuja integraali

$$ \int_\Omega f(x_1, \dots, x_m)\ \mathrm{d}x_1 \dots \mathrm{d}x_m $$

väärtus.

Idee on vaadelda piirkonda $\Omega$ kastis

$$ B = [\alpha_1, \beta_1] \times \dots \times [\alpha_m, \beta_m], $$

kus $\Omega \subset B$. Defineerime hulgal $B$ uue funktsiooni $F$ järgmiselt:

$$ F(x_1, \dots, x_m) = \begin{cases}
    f(x_1, \dots, x_m) & \text{ kui } (x_1, \dots, x_m) \in \Omega, \\
    0 & \text{ muudel juhtudel}.
\end{cases} $$

Siis saame integraali

$$ \int_\Omega f(x_1, \dots, x_m)\ \mathrm{d}x_1 \dots \mathrm{d}x_m $$

arvutada ligikaudselt järgmiselt:

$$ \int_\Omega f(x_1, \dots, x_m)\ \mathrm{d}x_1 \dots \mathrm{d}x_m \approx \frac{V}{N} \sum_{i=1}^{N} F(x_1^{(i)}, \dots, x_m^{(i)}), $$

kus $V$ tähistab kasti $B$ ruumala ning $x_1^{(i)}, \dots, x_m^{(i)}$ on $i = 1, \dots, n$ ja $j = 1, \dots, m$ jaoks sõltumatud ühtlaselt jaotatud juhuslikud arvud.

Seda arutelu realiseerib järgmine funktsioon.

```python
def mc_integrate(func, omega, B, N=100000):
    rs = np.array([a + (b - a)*random(N) for a, b in B])
    s = 0
    for i in range(N):
        r = rs[:,i]
        if omega(r):
            s += func(r)
    V = np.prod([b-a for a, b in B])
    return V*s/N
```

```python
omega = lambda x: True
f = lambda x: x[0]*x[1]**2
B = np.array([[0, 2], [0, 1]])

Math(r'\iint_{\Omega} xy^2\ \mathrm{d}x \mathrm{d}y \approx %.16f, \text{ kus } \Omega = [0, 2] \times [0, 1]' %
     mc_integrate(f, omega, B))
```
